```python
# Instalar si no lo tenés
# pip install yfinance pandas numpy

import yfinance as yf
import pandas as pd
import numpy as np

# 1️⃣ Descargar datos históricos
# 5 años de precios diarios
tickers = ["AAPL", "SPY"]
data = yf.download(tickers, start="2018-10-01", end="2026-10-01", auto_adjust=False,multi_level_index=False,progress=False)['Adj Close']

# 2️⃣ Calcular retornos diarios logarítmicos
returns = np.log(data / data.shift(1)).dropna()

# 3️⃣ Calcular beta usando covarianza / varianza del mercado
cov_matrix = returns.cov()  # matriz de covarianza diaria
beta_aapl = cov_matrix.loc["AAPL", "SPY"] / cov_matrix.loc["SPY", "SPY"]

print(f"Beta diaria de AAPL vs SPY: {beta_aapl:.3f}")

# 4️⃣ (Opcional) Beta anualizada, aunque beta no se anualiza, para comparación con R^2 o volatilidades:
# beta_aapl_anual = beta_aapl  # no se multiplica por 252

# 5️⃣ (Opcional) Regresión lineal para obtener intercepto y R^2
import statsmodels.api as sm
X = returns["SPY"]
y = returns["AAPL"]
X = sm.add_constant(X)  # agrega intercepto
model = sm.OLS(y, X).fit()
print(model.summary())
```
